```python
y = np.array(tf.keras.utils.to_categorical(y, num_classes=total_words))
y[0]
```

```python
model = Sequential()
model.add(Embedding(input_dim=total_words, output_dim=100, input_length=max_sequence_len-1))
model.add(LSTM(units = 150))
model.add(Dense(units = total_words, activation='softmax'))
print(model.summary())
```

```python
# No. of parameters in Embedding layer is equal to no. of dimensions x vocab size
# vocab_size = 209
# No. of dimensions = 100
# Params = 209x100 = 20900
# 
# No. of params in LSTM = 4×(units×(input_dim+units+1))
# units is the number of LSTM units in the layer.
# input_dim is the dimensionality of the input to the LSTM layer.
# 
# No. of params in LSTM = 4 x (150 x (100 + 150 + 1))
#                       = 4 x (150 x 251)
#                       = 4 x 37650
#                       = 150600
#                       
# Number of Parameters in Dense layer = (input_units+1)×output_units
#                                     = (150 + 1)x209
#                                     = 151 x 209
#                                     = 31,559
#                                     
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=100, verbose=1)
```

```python
input_text = "India needed"
predict_next_words= 10

for _ in range(predict_next_words):
    token_list = mytokenizer.texts_to_sequences([input_text])[0]
    print(token_list)
    token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')
    predicted = np.argmax(model.predict(token_list), axis=-1)
    output_word = ""
    for word, index in mytokenizer.word_index.items():
        if index == predicted:
            output_word = word
            break
    input_text += " " + output_word

print(input_text)
```
